```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def dec_to_binary(d):
    bnr=np.binary_repr(d,8)
    return bnr
```

```python
image=cv.imread("nwu.tif",cv.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
print(image)
```

```python
f = open("nwu.txt", "a")
for row in image:
    for i in row:
        f.write(dec_to_binary(i)+"\n")    
f.close()   
```
